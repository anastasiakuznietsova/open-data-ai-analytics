```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('nuclear_safety_q4_2025.xlsx')
```

Для окремого аналізу колонок можна прибрати 'year', 'quarter', 'station', 'iodine_radionuclides_index', 'stable_ radionuclides_index'. Індекс викидів радіонуклідів йоду та Індекс викидів ДІН - прибираємо, так як для всіх станцій у всі роки стоїть одне не числове значення "<0.01"

```python
df_num = df.drop(['year', 'quarter', 'station', 'iodine_ radionuclides_index', 'stable_ radionuclides_index'], axis=1)
for col in df_num.columns:
  fig, axes = plt.subplots(1, 2, figsize=(15, 4))
  sns.histplot(df_num[col], bins=30, kde=True, color='cornflowerblue', ax=axes[0])
  axes[0].set_title(f'Гістограма для колонки {col}')
  axes[0].set_xlabel(col)
  axes[0].set_ylabel('Частота')

  sns.boxplot(x=df_num[col], color='yellow', ax=axes[1])
  axes[1].set_title(f'Box Plot колонки {col}')
  axes[1].set_ylabel(col)
  plt.tight_layout()
  plt.show()
  print()
```

```python
df_boxplot = df.drop(['stable_ radionuclides_index','iodine_ radionuclides_index', 'station','quarter','year'],axis=1)
for col in df_boxplot.columns:
    plt.figure(figsize=(10, 5))

    sns.boxplot(data=df, x=col, y='station', hue='station', palette='Set2', legend=False)

    plt.title(f'Розподіл значень: {col} у розрізі АЕС', fontsize=12, fontweight='bold')
    plt.xlabel('Значення викидів', fontsize=10)
    plt.ylabel('Атомна електростанція', fontsize=10)

    plt.grid(axis='x', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

Перевіримо гіпотезу 1. про те, що ЗАЕС матиме найбільший індекс скиду серед всіх станцій, так як має найбільші блоки.

```python
plt.figure(figsize=(10, 5))

sns.barplot(data=df, x='station', y='index_dump', palette='viridis', hue='station', legend=False)

plt.title(f'Розподіл значень: index_dump у розрізі АЕС', fontsize=12, fontweight='bold')
plt.ylabel('Значення викидів', fontsize=10)
plt.xlabel('Атомна електростанція', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
